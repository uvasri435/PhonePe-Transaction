```python
# creating a dataframe for Top transaction
import pandas as pd
import json
import os

# Set the path correctly
path = "C:/Users/uva61/OneDrive/Desktop/project/pulse-master/data/top/transaction/country/india/state/"
Agg_state_list = os.listdir(path)

# Initialize the collection dictionary
clm = {'State': [], 'Year': [], 'Quarter': [], 'entityName': [], 'type': [], 'count': [], 'amount': []}

# Loop through each state
for state in Agg_state_list:
    state_path = os.path.join(path, state)
    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)
        for file in os.listdir(year_path):
            if file.endswith(".json"):
                file_path = os.path.join(year_path, file)
                try:
                    with open(file_path, 'r') as f:
                        data = json.load(f)

                        if 'data' in data:
                            for entity_type in ['states', 'districts', 'pincodes']:
                                entity_data = data['data'].get(entity_type)

                                # Proceed only if entity_data is a non-empty list
                                if isinstance(entity_data, list) and entity_data:
                                    for entity in entity_data:
                                        entity_name = entity.get('entityName')
                                        metric = entity.get('metric', {})
                                        count = metric.get('count')
                                        amount = metric.get('amount')

                                        if entity_name and count is not None and amount is not None:
                                            clm['State'].append(state)
                                            clm['Year'].append(year)
                                            clm['Quarter'].append(int(file.replace('.json', '')))
                                            clm['entityName'].append(entity_name)
                                            clm['type'].append(entity_type)
                                            clm['count'].append(count)
                                            clm['amount'].append(amount)
                                else:
                                    # No warning if empty or not list, just skip silently
                                    continue
                        else:
                            print(f"Warning: 'data' key not found in file {file_path}")

                except (FileNotFoundError, json.JSONDecodeError, KeyError, TypeError) as e:
                    print(f"Error processing file {file_path}: {e}")

# Create final DataFrame
Top_Trans = pd.DataFrame(clm)
```

```python
#Transaction Analysis Across States and Districts
import pandas as pd
from sqlalchemy import create_engine

# MySQL connection details
user = 'root'
password = 'uvasripradeep'
host = '127.0.0.1'
database = 'Phonepe'

# Create SQLAlchemy engine
engine = create_engine(f'mysql+mysqlconnector://{'root'}:{'uvasripradeep'}@{'127.0.0.1'}/{'Phonepe'}')


# SQL query to get top 10 districts by transaction amount
query = """
SELECT 
    entityName AS District,
    State,
    SUM(count) AS Total_Transactions,
    SUM(amount) AS Total_Amount
FROM top_transaction
WHERE type = 'districts'
GROUP BY entityName, State
ORDER BY Total_Amount DESC Limit 10;

"""

# Run the query and fetch result as DataFrame
df_top_districts = pd.read_sql(query, engine)

# Export to Excel
df_top_districts.to_excel("top_10_districts_by_transaction_amount.xlsx", index=False)
print("Top districts exported to Excel.")
```

```python
#Top districts by insurance transactions
import pandas as pd

# Set specific year and quarter
year = '2022'
quarter = 4

# MySQL connection details
user = 'root'
password = 'uvasripradeep'
host = '127.0.0.1'
database = 'Phonepe'

# Create SQLAlchemy engine
engine = create_engine(f'mysql+mysqlconnector://{'root'}:{'uvasripradeep'}@{'127.0.0.1'}/{'Phonepe'}')

# Top 10 districts by insurance transaction count and amount for a given year-quarter
query = f"""
SELECT 
    entityName AS District,
    State,
    SUM(count) AS Total_Insurance_Transactions,
    SUM(amount) AS Total_Insurance_Amount
FROM top_insurance

WHERE type = 'districts'
  AND Year = '{year}'
  AND Quarter = {quarter}
GROUP BY entityName, State
ORDER BY Total_Insurance_Transactions DESC
LIMIT 10;
"""

# Execute the query and load result into DataFrame
df_top_districts = pd.read_sql(query, engine)

# Export the results to Excel
df_top_districts.to_excel("top_districts_insurance_transactions.xlsx", index=False)
print("Top districts by insurance transactions exported successfully.")
# Load the Excel file back into a DataFrame
df = pd.read_excel("top_districts_insurance_transactions.xlsx")

# Display the data
print(df)  
```
